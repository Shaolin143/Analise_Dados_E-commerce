```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/ecomerce.csv')

df['faturamento'] = df['quantidade'] * df['preco']

df.head()
```

```python
faturamento_total = df['faturamento'].sum()
print(f'O faturamento total da empresa foi de R${faturamento_total}')
```

```python
top_produto = df.groupby('produto')['faturamento'].sum().sort_values(ascending=False)
nome_produto = top_produto.idxmax()
print(f'O produto com o maior faturamento foi {nome_produto}, com faturamento total de R${top_produto.max()}')
```

```python
mais_vendido = df.groupby('produto')['quantidade'].sum()
print(f'O produto mais vendido foi {mais_vendido.idxmax()}, com um total de {mais_vendido.max()} unidades vendidas.')
```

```python
melhor_cliente = df.groupby('cliente')['faturamento'].sum().sort_values(ascending=False)
nome_top_cleinte = melhor_cliente.idxmax()
receita_top_cliente = melhor_cliente.max()
print(f'{nome_top_cleinte} foi o cliete que gerou mais receita com um valor total de R${receita_top_cliente}')

```

```python
estado_maior_faturamento = df.groupby('estado')['faturamento'].sum().sort_values(ascending=False)
print(f'O estado que se destaca com o maior faturamento é {estado_maior_faturamento.idxmax()} que teve um faturamento total de R${estado_maior_faturamento.max()}')
```

```python
faturamento_categoria = df.groupby('categoria')['faturamento'].sum().sort_values(ascending=False)
print(f'{faturamento_categoria}\nFaturamento total: {faturamento_total}')
```
